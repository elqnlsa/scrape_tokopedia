```python
import re
import nltk
from collections import Counter
from nltk.corpus import stopwords

# --- download stopwords bahasa Indonesia (sekali aja perlu jalanin) ---
nltk.download("stopwords")

# --- ambil stopwords Indonesia ---
stopwords_id = set(stopwords.words("indonesian"))

# --- kolom yang akan dianalisis ---
cols_to_check = [
    "product_id", "product_name", "description", 
    "price", "sold_count", "rating", "total_stock", "category"
]

def get_most_common_words(series, n=10):
    text = " ".join(series.dropna().astype(str))
    words = re.findall(r'\w+', text.lower())
    
    filtered = []
    for w in words:
        # skip angka murni
        if w.isdigit():
            continue
        # skip stopwords
        if w in stopwords_id:
            continue
        filtered.append(w)
    
    return Counter(filtered).most_common(n)

# --- loop tiap kolom ---
for col in cols_to_check:
    print(f"\n=== Most Common Words in '{col}' ===")
    common_words = get_most_common_words(df[col], n=10)
    for word, count in common_words:
        print(f"{word}: {count}")
```

```python
# --- kumpulkan semua kata ---
all_words = []
for col in cols_to_check:
    common_words = get_most_common_words(df[col], n=None)  # ambil semua kata
    all_words.extend(common_words)

# --- gabungkan count dari semua kolom ---
total_counter = Counter()
for word, count in all_words:
    total_counter[word] += count

# --- ambil top 300 ---
top_words = total_counter.most_common(500)

# --- jadikan DataFrame ---
df_top_words1 = pd.DataFrame(top_words, columns=['word', 'count'])
df_top_words1.head
```

```python
import re
import pandas as pd

# slang Indonesia yang umum (bisa ditambah lagi)
SLANG_SEED = {
    # ketawa
    "wkwk","wk","wkwkwk","hahaha","hehe","huhu","hadeh",
    # kasar / makian
    "anjay","anjir","anjirr","anj","asu","goblok","tolol","coeg","cuk","tai","taii",
    # singkatan
    "cmiiw","btw","lol","gw","lu","loe","gmn","gk","ga","nggak","udh","udah","yg",
    "brb","afk","idk","irl","imo","imo","rip","fyi","otw",
    # sapaan
    "bro","sist","cuy","cuyy","ciy","gan","bosque","juragan",
    # gaya
    "mantul","mager","gabut","alay","santuy","keren","cihuy","asyik","asik","yoi","kuy"
}

SLANG_RULES = [
    # ketawa berulang
    ("repeated_laugh", re.compile(r"^(?:ha|he|hi|hu|wk|wkwk|x){2,}$")),

    # elongated huruf akhir (cuyyy, brooo, mantappp)
    ("elongated", re.compile(r"^[a-z]{2,}([a-z])\1{2,}$")),

    # slang angka â†’ huruf (4l4y, g4nt3ng, 3aa)
    ("leet_slang", re.compile(r"^(?=.*[0-9])(?=.*[a-z])[a-z0-9]{3,}$")),

    # whitelist slang pendek (harus ada di SLANG_SEED)
    ("seeded", re.compile(r"^[a-z]{2,10}$")),
]
def detect_slangs(df_words: pd.DataFrame, min_count: int = 5) -> pd.DataFrame:
    rows = []
    for _, row in df_words.iterrows():
        w = str(row["word"]).lower()
        c = int(row["count"])
        
        if c < min_count:
            continue
        if not re.fullmatch(r"[a-z0-9]+", w):
            continue

        matched = None
        for name, rx in SLANG_RULES:
            if rx.match(w):
                # kalau rule "seeded", cek dulu harus ada di SLANG_SEED
                if name == "seeded" and w not in SLANG_SEED:
                    continue
                matched = name
                break

        if not matched:
            continue

        rows.append({"word": w, "count": c, "rule": matched})

    return (
        pd.DataFrame(rows)
        .sort_values(["count","word"], ascending=[False, True])
        .reset_index(drop=True)
    )
    
# contoh pakai:
df_slangs = detect_slangs(df_top_words_combined, min_count=5)
print(df_slangs.head(40))
```
