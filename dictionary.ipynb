```python
import re
import nltk
from collections import Counter
from nltk.corpus import stopwords

# --- download stopwords bahasa Indonesia (sekali aja perlu jalanin) ---
nltk.download("stopwords")

# --- ambil stopwords Indonesia ---
stopwords_id = set(stopwords.words("indonesian"))

# --- kolom yang akan dianalisis ---
cols_to_check = [
    "product_id", "product_name", "description", 
    "price", "sold_count", "rating", "total_stock", "category"
]

def get_most_common_words(series, n=10):
    text = " ".join(series.dropna().astype(str))
    words = re.findall(r'\w+', text.lower())
    
    filtered = []
    for w in words:
        # skip angka murni
        if w.isdigit():
            continue
        # skip stopwords
        if w in stopwords_id:
            continue
        filtered.append(w)
    
    return Counter(filtered).most_common(n)

# --- loop tiap kolom ---
for col in cols_to_check:
    print(f"\n=== Most Common Words in '{col}' ===")
    common_words = get_most_common_words(df[col], n=10)
    for word, count in common_words:
        print(f"{word}: {count}")
```

```python
# --- kumpulkan semua kata ---
all_words = []
for col in cols_to_check:
    common_words = get_most_common_words(df[col], n=None)  # ambil semua kata
    all_words.extend(common_words)

# --- gabungkan count dari semua kolom ---
total_counter = Counter()
for word, count in all_words:
    total_counter[word] += count

# --- ambil top 300 ---
top_words = total_counter.most_common(500)

# --- jadikan DataFrame ---
df_top_words1 = pd.DataFrame(top_words, columns=['word', 'count'])
df_top_words1.head
```

```python
import re
import pandas as pd

# === whitelist slang Indonesia (lebih banyak) ===
SLANG_SEED = {
    # ketawa / ekspresi
    "wkwk", "wk", "wkwkwk", "hahaha", "hehe", "huhu", "hadeh", "hahaa", "heh", "hihi",
    "huehue", "wkwkw", "hehehe", "hihihi",

    # sapaan / panggilan gaul
    "bro", "brooo", "cuy", "cuyyy", "ciy", "sisss", "gan", "ganzz", "mblooo", "sis", "bosque", "juragan",

    # gaya / slang umum
    "mantul", "mantulll", "mager", "gabut", "alay", "alayyyy", "santuy", "santuyyy", 
    "keren", "kerenz", "cihuy", "cihuyyy", "asyik", "asyikkk", "asik", "yoi", "kuy", 
    "sokap", "parahh", "goks", "mantep", "sip", "ok", "wow", "hah", "yaa", "wkwkw",

    # kata populer sosial media
    "baper", "galau", "nyantai", "kepo", "recehh", "pede", "gas", "gass", "hadehh", "hadehhh"
}

SLANG_RULES = [
    ("repeated_laugh", re.compile(r"^(?:ha|he|hi|hu|wk|wkwk|x){2,}$")),
    ("elongated", re.compile(r"^[a-z]{2,}([a-z])\1{2,}$")),
    ("seeded", re.compile(r"^[a-z]{2,15}$")),  # diperluas jadi 15 huruf
]

def detect_slangs(df_words: pd.DataFrame, min_count: int = 5) -> pd.DataFrame:
    rows = []
    for _, row in df_words.iterrows():
        w = str(row["word"]).lower()
        c = int(row["count"])

        if c < min_count:
            continue
        if not re.fullmatch(r"[a-z0-9]+", w):
            continue

        matched = None
        for name, rx in SLANG_RULES:
            if rx.match(w):
                if name == "seeded" and w not in SLANG_SEED:
                    continue
                matched = name
                break

        if matched:
            rows.append({"word": w, "count": c, "rule": matched})

    return pd.DataFrame(rows).sort_values(["count","word"], ascending=[False, True]).reset_index(drop=True)

# contoh pakai:
df_slangs = detect_slangs(df_top_words_combined, min_count=20)
df_slangs.head(40)
```

```python
import pandas as pd

INDO_CONTRACTIONS = {
    "gk": "tidak",
    "gak": "tidak",
    "nggak": "tidak",
    "ga": "tidak",
    "yg": "yang",
    "dgn": "dengan",
    "dr": "dari",
    "sdh": "sudah",
    "blm": "belum",
    "sbg": "sebagai",
    "tp": "tapi",
    "klo": "kalau",
    "klu": "kalau",
    "krn": "karena",
    "dll": "dan lain-lain",
    "dst": "dan seterusnya"
}

def detect_indo_contractions_count(df_words: pd.DataFrame) -> pd.DataFrame:
    """
    Mendeteksi contractions Bahasa Indonesia dari df_words,
    menghitung jumlah kemunculannya.
    df_words: DataFrame dengan kolom 'word' (token)
    """
    counter = {}
    for w in df_words["word"]:
        token = str(w).lower()
        if token in INDO_CONTRACTIONS:
            counter[token] = counter.get(token, 0) + 1
    
    rows = [{"word": k, "count": v, "rule": "indo_contraction"} for k, v in counter.items()]
    return pd.DataFrame(rows).sort_values(["count", "word"], ascending=[False, True]).reset_index(drop=True)

df_indo_contractions = detect_indo_contractions_count(df_top_words_combined)
df_indo_contractions.head(40)
```
